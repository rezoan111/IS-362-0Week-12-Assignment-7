```python
import requests
import pandas as pd
```

```python
# My NYT API key
api_key = "MY_NYT_API_KEY"
```

```python
# This cell connects to the NYT Most Popular API
url = "https://api.nytimes.com/svc/mostpopular/v2/viewed/1.json"

params = {
    "api-key": api_key
}

response = requests.get(url, params=params)

# Check if the request worked
print(response.status_code)
```

```python
# This cell reads the JSON data from the API
data = response.json()

# Show the main keys in the JSON
print(data.keys())

# Show how many articles were returned
print(len(data["results"]))
```

```python
# This cell turns the NYT results into a pandas DataFrame
df = pd.DataFrame(data["results"])

# Show the column names
print(df.columns.tolist())

# Show the first 5 rows
df.head()
```

```python
# This cell keeps only the columns I want to show clearly
df_clean = df[[
    "title",
    "published_date",
    "section",
    "byline",
    "source",
    "url"
]]

# Show the cleaned DataFrame
df_clean.head()
```

```python
# This cell shows the size of my cleaned DataFrame
print("Rows and columns:", df_clean.shape)

# This cell shows the data types
print(df_clean.dtypes)
```
